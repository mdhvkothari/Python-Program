```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data= pd.get_dummies(data)
```

```python
train = data[0:7999]
test = data[8000:]
```

```python
x_train = train.drop('Item_Outlet_Sales',axis = 1)
```

```python
x_train
```

```python
y_train = train['Item_Outlet_Sales']
```

```python
x_test = test.drop('Item_Outlet_Sales',axis = 1)
```

```python
true_p = test['Item_Outlet_Sales']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#now linear regression does not accept string value so we have to convert them into dummies values
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
```

```python
x_train.fillna(0,inplace=True)
```

```python
x_test.fillna(0,inplace=True)
```

```python
linerRegession = LinearRegression()
```

```python
linerRegession.fit(x_train,y_train)
```

```python
pred = linerRegession.predict(x_test)
```

```python
linerRegession.score(x_test,true_p)
```

```python
np.sqrt(np.mean(np.power((np.array(true_p)-np.array(pred)),2)))
```
